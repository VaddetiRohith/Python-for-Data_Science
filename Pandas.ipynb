# Series
- series is a one dimenstional labled array and capable of holding data of any type

```python
lst = [10,10.5,'Hello']
lst
```

```python
# Creating Series One way

s = pd.Series(lst)
print(s)
type(s)
```

```python
# Creating Series Second way

s1 = pd.Series(data=['Python', 'Data Science', 'DL'], index=['Subject_1', 'Subject_2', 'Subject_3'])
s1
```

```python
# Creating Seies Third way

d = {1 : 'Python', 2: 'Data Science', 3: 'DL'}
print(d)
print(type(d))

print('---- Converting dict to Series ----')

d = pd.Series(d)
print(d)
print(type(d))
```

```python
#Accessing elements from a series using index lable

s1['Subject_1']
```

```python
#Accessing element from series Using index value
s1[0]
```

```python
s1[1:]
```
